```python
folder_path = os.getcwd()
files = os.listdir(folder_path)

excel_files = [file for file in files if file.endswith('.xlsx') or file.endswith('.xls')]
excel_files_sort = sorted(excel_files)


cmap = plt.cm.Reds
norm = plt.Normalize(vmin=0, vmax=50)

def filter_data(df, Cycle_index):
    filtered_df = df[df['Cycle Index'].isin([Cycle_index])]
    filtered_df.reset_index(drop=True, inplace=True)
    voltage_data = filtered_df['Voltage(V)']
    capacity_data = filtered_df['Capacity(Ah)']
    
    return voltage_data, capacity_data

for file in excel_files_sort:
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path, header=0, sheet_name='record')

    # Plotting Cycle Index 1 separately
    for Cycle_index in [1]:
        voltage_data, capacity_data = filter_data(df, Cycle_index)
        plt.plot(capacity_data, voltage_data, linestyle='--', linewidth=1, label='Charge Cycle 1', color='black')
    
    # Plotting Cycle Index 2 to 21
    for Cycle_index in range(2, 50):
        voltage_data, capacity_data = filter_data(df, Cycle_index)
        color = cmap(norm(Cycle_index**10))
        plt.plot(capacity_data, voltage_data, linestyle='--', linewidth=1, color=color)
        # plt.plot(capacity_data, voltage_data, linestyle='--', linewidth=1, label=f'Charge Cycle {Cycle_index}', color=color)

    
    plt.grid()
    plt.legend()
    plt.show()
```
